How does `matplotlib` decides the bins? Actually `matplotlib`'s `hist` function uses `numpy`'s `histogram` function under the hood.

- matplotlib.pyplot.hist() calls numpy.histogram() internally to calculate the bins for the histogram.
- If numpy.histogram() is used without specifying the number of bins, it determines the optimal number of bins based on the data distribution. However, if we use matplotlib.pyplot.hist() without specifying the bins parameter, it defaults to 10 bins.
- We can control the bin selection using options like 'auto', 'sturges', 'fd', etc., which yield different outcomes depending on the dataset size and characteristics.
    - The 'auto' strategy selects a suitable number of bins based on the data distribution and may use other rules such as 'sturges' or 'fd' depending on the data characteristics.
    - If bins is omitted in matplotlib.pyplot.hist(), it will use a default of 10 bins.

Do you see the issues with having too few bins or too many bins? In particular, do you notice weird patterns that emerge from `bins=30`?
- When the number of bins is too small, the histogram becomes overly simplified. It cannot accurately reflect the distribution's detailed characteristics. For example, bins=2 or bins=3 only capture the most basic trend in the data, which might mask important details like peaks or irregularities.
- With a very high number of bins, the histogram becomes too detailed, and the bars represent very narrow ranges. The result can be noisy, with many individual spikes that may not represent meaningful patterns but instead random variations in the data.
- When using bins=30, I can notice patterns such as small peaks and troughs that appear to fluctuate. These patterns could be artifacts caused by the interaction between the exact data points and the bin boundaries. Sometimes, the choice of bin size can emphasize certain minor variations or outliers, making them appear more important than they really are.

**Q: Can you guess why do you see such patterns? What are the peaks and what are the empty bars? What do they tell you about choosing the binsize in histograms?**

```python
# YOUR SOLUTION HERE

plt.hist(movies.IMDB_Rating, bins=60)
plt.show()
```

```python
# YOUR SOLUTION HERE

n, bins, patches = plt.hist(movies.IMDB_Rating, bins=60)
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.show()

peaks = []
empty_bars = []

# Top 20% bins as a threshold for peak detection 
peak_threshold = np.percentile(n, 80)

for i, count in enumerate(n):
    if count > peak_threshold:
        peaks.append((i + 1, bins[i], bins[i+1], count))  # Bin number, range, count
    elif count == 0:
        empty_bars.append((i + 1, bins[i], bins[i+1]))  # Bin number, range

print("Peaks (Bin number, Range, Count):")
for peak in peaks:
    print(f"    Bin {peak[0]} (Range: {peak[1]} - {peak[2]}) has {peak[3]} movies.")

print("\nEmpty Bars (Bin number, Range):")
for empty_bar in empty_bars:
    print(f"    Bin {empty_bar[0]} (Range: {empty_bar[1]} - {empty_bar[2]}) is empty.")
```

- Peaks: Information about bins that contain the most ratings, helping to identify ranges where movies are densely populated.
- Empty Bars: Information about bins with zero ratings, highlighting gaps in the data.

**Why do you see such patterns?**

The patterns you observe in the histogram with bins=60 arise due to the interaction between the discrete nature of the IMDB_Rating data and the chosen bin size. Ratings are typically given in whole numbers or increments like 0.5, leading to clusters at these specific values. When the bin size does not align well with these common rating points, it creates alternating peaks and empty bars. Peaks form where the ratings cluster, while empty bars appear where there are few or no ratings, often due to misaligned bin boundaries and overly fine granularity.

**What are the peaks and what are the empty bars?**

The peaks in the histogram represent ranges of ratings where many movies share similar scores, indicating clusters at popular ratings. These peaks show which ratings are most common among the movies. In contrast, the empty bars represent bins that contain very few or no movies, indicating that the corresponding rating range is either uncommon or entirely unrepresented. This uneven distribution highlights the fact that movie ratings are not evenly spread across all possible values but are instead concentrated at certain popular points.

**What do they tell you about choosing the bin size in histograms?**

The presence of pronounced peaks and empty bars suggests that the chosen bin size has a significant impact on how well the histogram represents the data distribution. A bin size that is too small leads to excessive noise and gaps between bins, while a bin size that is too large can oversimplify the data. For discrete data like movie ratings, it seems to be crucial to select a bin size that aligns well with the rating intervals to accurately reflect the clustering of the data and avoid creating misleading patterns in the histogram.

## Formulae for choosing the number of bins.

We can manually choose the number of bins based on those formulae.

```python
N = len(movies)

plt.figure(figsize=(12,4))

# Sqrt
nbins = int(np.sqrt(N))

plt.subplot(1,3,1)
plt.title("SQRT, {} bins".format(nbins))
movies.IMDB_Rating.hist(bins=nbins)

# Sturge's formula
nbins = int(np.ceil(np.log2(N) + 1))

plt.subplot(1,3,2)
plt.title("Sturge, {} bins".format(nbins))
movies.IMDB_Rating.hist(bins=nbins)

# Freedman-Diaconis
iqr = np.percentile(movies.IMDB_Rating, 75) - np.percentile(movies.IMDB_Rating, 25)
width = 2*iqr/np.power(N, 1/3)
nbins = int((max(movies.IMDB_Rating) - min(movies.IMDB_Rating)) / width)

plt.subplot(1,3,3)
plt.title("F-D, {} bins".format(nbins))
movies.IMDB_Rating.hist(bins=nbins)
```

But we can also use built-in formulae too. Let's try all of them.

```python
plt.figure(figsize=(20,4))

plt.subplot(161)
movies.IMDB_Rating.hist(bins='fd')

plt.subplot(162)
movies.IMDB_Rating.hist(bins='doane')

plt.subplot(163)
movies.IMDB_Rating.hist(bins='scott')

plt.subplot(164)
movies.IMDB_Rating.hist(bins='rice')

plt.subplot(165)
movies.IMDB_Rating.hist(bins='sturges')

plt.subplot(166)
movies.IMDB_Rating.hist(bins='sqrt')
```

Some are decent, but several of them tend to overestimate the good number of bins. As you have more data points, some of the formulae may overestimate the necessary number of bins. Particularly in our case, because of the precision issue, we shouldn't increase the number of bins too much.

### Then, how should we choose the number of bins?

So what's the conclusion? use Scott's rule or Sturges' formula?

No, I think the take-away is that you **should understand how the inappropriate number of bins can mislead you** and you should **try multiple number of bins** to obtain the most accurate picture of the data. Although the 'default' may work in most cases, don't blindly trust it! Don't judge the distribution of a dataset based on a single histogram. Try multiple parameters to get the full picture!

As mentioned before, in `altair` histogram is not special. It is just a plot that use bars (`mark_bar()`) where X axis is defined by `IMDB_Rating` with bins (`bin=True`), and Y axis is defined by `count()` aggregation function.

```python
alt.Chart(vega_datasets.data.movies.url).mark_bar().encode(
    alt.X("IMDB_Rating:Q",  bin=True),
    alt.Y('count()')
)
```

Have you noted that it is `IMDB_Rating:Q` not `IMDB_Rating`? This is a shorthand for

```python
alt.Chart(vega_datasets.data.movies.url).mark_bar().encode(
    alt.X('IMDB_Rating', type='quantitative', bin=True),
    alt.Y(aggregate='count', type='quantitative')
)
```

In altair, you want to specify the data types using one of the four categories: quantitative, ordinal, nominal, and temporal. https://altair-viz.github.io/user_guide/encoding.html#data-types

Although you can adjust the bins in `altair`, it does not encourage you to set the bins directly. For instance, although there is `step` parameter that directly sets the bin size, there are parameters such as `maxbins` (maximum number of bins) or `minstep` (minimum allowable step size), or `nice` (attemps to make the bin boundaries more human-friendly), that encourage you not to specify the bins directly.

```python
from altair import Bin

alt.Chart(vega_datasets.data.movies.url).mark_bar().encode(
    alt.X("IMDB_Rating:Q",  bin=Bin(step=0.09)),
    alt.Y('count()')
)
```

```python
alt.Chart(vega_datasets.data.movies.url).mark_bar().encode(
    alt.X("IMDB_Rating:Q",  bin=Bin(nice=True, maxbins=20)),
    alt.Y('count()')
)
```

### Composing charts in altair

Shall we avoid some repetitions? You can define a *base* empty chart first and then assign encodings later when you put together multiple charts together. Here is an example: https://altair-viz.github.io/user_guide/compound_charts.html#repeated-charts

**Q: Using the base chart approach to create a 2x2 chart where the top row shows the two histograms of `IMDB_Rating` with `maxbins`=10 and 50 respectively, and the bottom row shows another two histograms of `IMDB_Votes` with `maxbins`=10 and 50.**

```python
# YOUR SOLUTION HERE

# Base chart definition using the movies dataset URL
movies = vega_datasets.data.movies.url

base = alt.Chart().mark_bar().encode().properties(
    width=300,
    height=150
)

chart = alt.vconcat(data=movies)

# Top row: Two histograms of IMDB_Rating with different maxbins
row1 = alt.hconcat()
for maxbins in [10, 50]:
    row1 |= base.encode(
        alt.X("IMDB_Rating:Q", bin=alt.Bin(maxbins=maxbins)),
        alt.Y('count()')
    ).properties(title=f"IMDB Rating with maxbins={maxbins}")

# Bottom row: Two histograms of IMDB_Votes with different maxbins
row2 = alt.hconcat()
for maxbins in [10, 50]:
    row2 |= base.encode(
        alt.X("IMDB_Votes:Q", bin=alt.Bin(maxbins=maxbins)),
        alt.Y('count()')
    ).properties(title=f"IMDB Votes with maxbins={maxbins}")

# Arrange the charts in a 2x2 grid
chart &= row1
chart &= row2

chart
```
